# Numpy 1. Cleaning up data

Our first work with `numpy` arrays will involve some maninpulation of the data.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We have done an experiment that generated a set of data. We have done the experiment 3 times, so we have 3 sets of data. By now, we will assume we have converted that into numpy arrays. In fact, for this example, we will generate the data and add some random noise to it

```python
x = np.linspace(0, 4*np.pi, 100)
noise = (np.random.random(len(x))-0.5)*0.7
y1 = np.sin(x)+ 1 + noise
noise = (np.random.random(len(x))-0.5)*0.7
y2 = np.sin(x)+ 1 + noise
noise = (np.random.random(len(x))-0.5)*0.7
y3 = np.sin(x)+ 1 + noise
```

Let's plot the data to see how it looks like:

```python
for y in (y1, y2, y3):
    plt.plot(x,y)
plt.hlines(0,0, 4*np.pi, lw=0.5)
```

First, average the data:

```python
y_ave = (y1+y2+y3)/3
```

Next,we want to smooth the data. There are much better algorithms (implemented in scipy and scikit-learn) but for now, we will do a simple running average. Each point will be the average of their 2 neighbours. We have to decide how to treat the end points. As a first simplification we will remove them.

```python
y_smooth = (y_ave[:-2]+y_ave[1:-1]+y_ave[2:])/3 
```

Imagine that for further processing we need the data to be positive. Because of the noise, we could have negative values. Let's check if there are negative values:

```python
np.any(y_smooth<0)
```

And let's set them to zero. Remember that a boolean array can set the indices:

```python
y_smooth[y_smooth<0] = 0.
np.any(y_smooth<0)
```

## Digression

The generation of the 3 sets of "experimental" data was simple but a bit verbose and repetititve. It also implies typing 3 names of different variables if we want to operate on them. A simpler way would be to collect the data (or directly generate it) in a 2 dimensional array. Here we'll see how to do that and how to work with these 2D arrays. 

```python
noise_all = noise = (np.random.random((3,len(x)))-0.5)*0.4
y_all = np.sin(x)+ 1 +noise # wow, broadcasting in action!
```

To average, we could explicitly name all 3 rows:

```python
y_ave1 =(y_all[0]+y_all[1]+y_all[2])/3
```

which is not nice. We should sum along the first dimension, but not along the second. By default, `np.sum` sums all dimensions, but if we look at the documentation (remember `np.sum?`) we can find our how to do this:

```python
y_ave2 = np.sum(y_all, axis=0)/3
```

Of course, it would have been simpler to directly find the mean. Let's look for something related to  "average":

```python
np.lookfor('average', import_modules=False)
```

```python
y_ave3 = np.mean(y_all, axis=0)
```

Let's check they all agree:

```python
np.all(y_ave1==y_ave2), np.all(y_ave2==y_ave3) 
```
